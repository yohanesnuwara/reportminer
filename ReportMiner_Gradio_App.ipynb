```python
import gradio as gr
import os
import numpy as np
from google.colab import drive
from reportminer import rag, rag_folder  # Updated import to use reportminer module

# Mount Google Drive
drive.mount('/content/drive')

def list_folders(drive_path):
    """List folders in the specified Google Drive directory."""
    folders = [
        f for f in os.listdir(drive_path) if os.path.isdir(os.path.join(drive_path, f))
    ]
    return folders

def list_documents(folder_path):
    """List all documents in the specified folder."""
    documents = [
        f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))
    ]
    return documents

def process_selected_folder(folder_path, rag_models):
    """Run the processing logic for the selected folder."""
    try:
        # List all documents in the folder
        documents = list_documents(folder_path)
        documents_list = "\n".join(documents)

        # Process embeddings of these PDFs
        # rag_folder.PDF_folder(folder_path, rag_models)  # Pass rag_models explicitly
        rag_folder.Process(folder_path, rag_models)  # Pass rag_models explicitly

        return f"Processing completed successfully.\n\nDocuments in folder:\n{documents_list}"
    except Exception as e:
        return f"An error occurred: {str(e)}"

def ask_question(question, rag_models):
    """Ask a question to the RAG model."""
    try:
        # Query the RAG model
        answer, document, page_img = rag_folder.Ask(question, rag_models)

        # Convert image to array
        page_img = np.array(page_img)
        return answer, document, page_img
    except Exception as e:
        return f"An error occurred while querying: {str(e)}", None, None

# Specify the base Google Drive path
base_drive_path = '/content/drive/My Drive'

# Get folders in Google Drive
folders = list_folders(base_drive_path)

# Gradio app interface
def gradio_interface():
    def process(folder_name):
        folder_path = os.path.join(base_drive_path, folder_name)
        return process_selected_folder(folder_path, rag_models)

    def query_model(question):
        answer, document, page_img = ask_question(question, rag_models)
        if page_img is not None:
            return f"Answer: {answer}\nSource Document: {document}", page_img
        return f"Answer: {answer}\nSource Document: {document}", None

    with gr.Blocks() as app:
        gr.Markdown("""
        # ReportMiner: A Solution for Your Technical Documents 📂
        """)

        # Folder processing section
        with gr.Row():
            with gr.Column():
                folder_dropdown = gr.Dropdown(
                    choices=folders,
                    label="Select Folder",
                    value=folders[0] if folders else None
                )
                process_button = gr.Button("Process")
            with gr.Column():
                output_display = gr.Textbox(label="Processing Output", lines=5)

        process_button.click(process, inputs=[folder_dropdown], outputs=[output_display])

        # Question asking section
        gr.Markdown("""
        ## Let DeepSeek-VL + ColPali assist your query 🧐
        Enter a question below to retrieve information from the processed documents.
        """)

        with gr.Column():
            question_input = gr.Textbox(label="Your Question", placeholder="Type your question here...")
            query_button = gr.Button("Ask")
        with gr.Row():
            with gr.Column():
                answer_display = gr.Textbox(label="Answer", lines=5)
            with gr.Column():
                page_image = gr.Image(label="Page Image")

        query_button.click(query_model, inputs=[question_input], outputs=[answer_display, page_image])

    return app


# Run the Gradio app
gradio_app = gradio_interface()
gradio_app.launch()
```
